```python
img_blue = (img[:,:,0] < 150) & (img[:,:,1] < 150) & (img[:,:,2] > 150)
plt.imshow(img_blue)
```

```python
img_blue = label(img_blue)
table_blue=regionprops_table(img_blue,properties=("centroid","area","bbox"))
table_blue=pd.DataFrame(table_blue)
```
